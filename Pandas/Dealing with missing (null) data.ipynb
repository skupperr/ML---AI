```python
# If we have a lot of missing data then removing rows or columns may not be preferable. In such cases data scientists use Imputation technique. Just a fancy way of saying, fill it with whatever value. A good guess would be filling missing values by the mean of the dataset.

import numpy as np

df = pd.read_csv('csv files/manual_vs_auto.csv')

df['Manual'] = df.apply(lambda row: (round((row['Auto_th_2']+row['Auto_th_3']+row['Auto_th_3'])/3)) if np.isnan(row['Manual']) else row['Manual'], axis=1)

print(df.head(25))
```
