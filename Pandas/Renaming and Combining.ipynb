## Combining

https://www.kaggle.com/datasets/singole/youtube-dataset-of-countries/data

```python
canadian_youtube = pd.read_csv("csv files/CAvideos.csv")
british_youtube = pd.read_csv("csv files/GBvideos.csv")

df_videos = pd.concat([canadian_youtube, british_youtube])
df_videos.head(2)
```

```python
# join() lets you combine different DataFrame objects which have an index in common. For example, to pull down videos that happened to be trending on the same day in both Canada and the UK, we could do the following:

left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK').head(3)
```
