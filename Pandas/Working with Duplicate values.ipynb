```python
# keep True vs False

import pandas as pd

data = {'Manual': ['A', 'B', 'A', 'C', 'B', 'D']}
df = pd.DataFrame(data)

# Marks all duplicates as True except for the first occurrence. The first occurrence is considered unique.
# keep='first' (default)
print(df[df.duplicated(subset=['Manual'], keep='first')])

# Marks all duplicates as True` except for the last occurrence. The last occurrence is considered unique.
# keep='last'
print(df[df.duplicated(subset=['Manual'], keep='last')])

#Marks all occurrences of duplicates as True. No occurrence is considered unique.
# keep=False
print(df[df.duplicated(subset=['Manual'], keep=False)])
```

The subset parameter in the duplicated() method specifies which columns to consider when identifying duplicates. Here's the difference between using subset and not using it:

1. Without subset (default):
If subset is not specified, the method considers all columns in the DataFrame to determine duplicates.
Rows are considered duplicates only if all column values are identical.
2. With subset:
If subset is specified, only the columns listed in subset are considered when identifying duplicates.
Rows are considered duplicates if the values in the specified columns are identical, regardless of the values in other columns.

```python
import pandas as pd

data = {
    'Manual': ['A', 'B', 'A', 'C', 'B', 'D'],
    'Auto': [1, 2, 1, 3, 4, 5]
}
df = pd.DataFrame(data)

# Without subset (considers all columns)
print(df[df.duplicated(keep=False)])

# With subset (considers only the 'Manual' column)
print(df[df.duplicated(subset=['Manual'], keep=False)])
```

```python
# To see duplicates according to a specific column:

duplicate_manuals = df[df.duplicated(subset=['Manual'], keep=False)]
duplicate_manuals
```
