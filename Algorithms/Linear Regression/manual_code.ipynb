```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt	

df = pd.read_csv('cells.csv')
```

```python
x_df = df['time']
x_df = x_df.to_numpy()

y_df = df['cells']
y_df = y_df.to_numpy()

plt.scatter(x_df, y_df, color='blue', label='data')	
```

```python
sum_of_Xs = np.sum(x_df)
sum_of_Ys = np.sum(y_df)

product_of_Xs_and_Ys = np.multiply(x_df, y_df)
sum_of_product_of_Xs_and_Ys = np.sum(product_of_Xs_and_Ys)

x_squared = x_df * x_df
sum_of_squared_X = np.sum(x_squared)

sum_of_Xs_squared = sum_of_Xs * sum_of_Xs

n = x_df.shape[0]
# Y = mX + c

m = (n * sum_of_product_of_Xs_and_Ys - sum_of_Xs * sum_of_Ys) / (n * sum_of_squared_X - sum_of_Xs_squared)
c = (sum_of_Ys - m * sum_of_Xs) / n

print(f"m = {m}, c = {c}")
```

```python
print(m*2.3 + c)
```

```python
plt.scatter(x_df, y_df, color='blue', label='data')	
plt.plot(x_df, m * x_df + c, color='red', label='linear regression')
```
