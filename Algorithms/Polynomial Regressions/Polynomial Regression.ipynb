```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df1 = pd.read_csv('linear_data.csv')
x_linear = df1[['x']]
y_linear = df1[['y']]
```

```python
df2 = pd.read_csv('nonlinear_data.csv')
x_nonlinear = df2[['x']]
y_nonlinear = df2[['y']]
```

## Visual Representation

```python
# linear
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Relationship')
plt.legend()

# non-linear
plt.subplot(1,2,2)
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with non Linear relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-Linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Correlations
df1.corr()
```

```python
df2.corr()
```

## Linear

```python
from sklearn.linear_model import LinearRegression

reg1 = LinearRegression()
reg1.fit(x_linear, y_linear)
```

```python
reg1.score(x_linear, y_linear)
```

```python
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear relationship')
plt.plot(x_linear, reg1.predict(x_linear), color='red', label='Best fit line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Relationship')
plt.legend()
```

## Non-Linear

```python
reg2 = LinearRegression()
reg2.fit(x_nonlinear , y_nonlinear)
```

```python
reg2.score(x_linear, y_linear)
```

```python
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.plot(x_nonlinear, reg2.predict(x_nonlinear), color='red', label='Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()
```

## Polynomial Regression


```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=6)
X_poly = poly.fit_transform(x_nonlinear)
X_poly.shape   # Features = 7
```

```python
reg_poly = LinearRegression()
reg_poly.fit(X_poly, y_nonlinear)
```

```python
reg_poly.score(X_poly, y_nonlinear)
```

## Seperate Train Test


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_poly, y_nonlinear, test_size=0.3, random_state=42)
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Training score
reg.score(x_train, y_train)
```

```python
# Testing score
reg.score(x_test, y_test)
```
